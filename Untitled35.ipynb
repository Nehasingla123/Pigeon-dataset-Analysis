```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('penguins.csv')
print(df)
```

```python
df.isnull().any()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
# drop null values
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
pd.get_dummies(df['sex'])
```

```python
sex = pd.get_dummies(df['sex'],drop_first=True)
print(sex)
```

```python
pd.get_dummies(df['island'])
```

```python
island = pd.get_dummies(df['island'],drop_first=True)
print(island)
```

```python
data = pd.concat([df,island,sex],axis=1)
print(data)
```

```python
print(data.head(5))
```

```python
data.drop(['sex','island'],axis=1,inplace=True)
```

```python
print(data.head(5))
```

```python
# use map function to convert categorical values into numeric
Y = data.species
```

```python
Y = Y.map({'Adelie': 0, 'Chinstrap': 1, 'Gentoo': 2})
```

```python
Y.head()
```

```python
data.drop('species', axis=1, inplace=True)
```

```python
X = data
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifer = RandomForestClassifier(n_estimators=5)
classifer.fit(X_train,Y_train)
```

```python
y_pred = classifer.predict(X_test)
```

```python
y_pred
```

```python
classifer.score(X_test,Y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifer = RandomForestClassifier(n_estimators=7, criterion='gini')
classifer.fit(X_train,Y_train)
```

```python
classifer.score(X_test,Y_test)
```
